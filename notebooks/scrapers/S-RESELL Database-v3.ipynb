```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.keys import Keys
```

```python
df = pd.read_csv('KOF_04152019+resale-1.csv')
```

```python
df = df[df['Resale'].isna()]
```

```python
shoes = list(df.Name)
```

```python
#update driver here: http://chromedriver.chromium.org/downloads
driver = webdriver.Chrome(executable_path='/Users/nikhilbhargava/Documents/Shtocks/Selenium/chromedriver')
```

```python
driver.get("https://stockx.com/")
```

```python
detail_price = []
detail_price_d = {}
avg_price = []
avg_price_d = {}

for i in range(len(shoes)):
    try:
        shoe = shoes[i]
        url = 'https://stockx.com/search?s=' + str(shoe)
        driver.get(url)
        driver.find_element_by_css_selector('#search-wrapper > div.search-results-grid > div:nth-child(1) > a > h4').click()
        #search-wrapper > div.search-results-grid > div:nth-child(1) > a > h4
        new_url = driver.current_url
        responseMain = requests.get(new_url)
        soupMain = BeautifulSoup(responseMain.text, 'html.parser')
        d_price = soupMain.findAll('script')[29]
        a_price = soupMain.findAll('div')[0]
                
        detail_price_d[shoe] = d_price
        detail_price.append(d_price)
        avg_price_d[shoe] = a_price
        avg_price.append(a_price)
        
    except:
        detail_price_d[shoe] = ''
        detail_price.append('')
        avg_price_d[shoe] = ''
        avg_price.append('')
    
    if (i%100==0):
        print('Finished scraping the', str(i) + 'th/st shoe.')  
```

```python
file = open("resale-detail-list-2.txt", "w") 
yes = str(detail_price)
file.write(yes)
file.close()
```

```python
file = open("resale-detail-dict-2.txt", "w") 
yes = str(detail_price_d)
file.write(yes) 
file.close()
```

```python
file = open("resale-avg-list-2.txt", "w") 
yes = str(avg_price)
file.write(yes) 
file.close()
```

```python
file = open("resale-avg-dict-2.txt", "w") 
yes = str(avg_price_d)
file.write(yes) 
file.close()
```

```python
magic_word = 'Average Sale Price</div><div class="gauge-value">'
magic_sift = len(magic_word)
```

```python
df_resale = []

for i in range(len(avg_price)):
    shoe_stock = avg_price[i]
    
    if shoe_stock == '':
        df_resale.append(None)
    else:
        price_block = shoe_stock
        text_pblock = str(price_block)
        finder = text_pblock.find(magic_word)
        if finder == -1:
            df_resale.append(None)
            continue
        init_pos = finder + magic_sift
        pos = init_pos
        
        extracted_avg_price = ''
        current_letter = text_pblock[init_pos]
        while current_letter != '<':
            if pos == init_pos:
                pos += 1
                current_letter = text_pblock[pos]
            else:
                extracted_avg_price += current_letter
                pos += 1
                current_letter = text_pblock[pos]
        df_resale.append(extracted_avg_price)
```

```python
df['Avg_Resale'] = df_resale
```
