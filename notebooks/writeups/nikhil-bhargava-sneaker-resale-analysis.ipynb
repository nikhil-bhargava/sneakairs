# Sneaker Resale Analysis: What causes some sneakers to increase in value after they're released?

As an avid "sneakerhead", I often find myself thwarted by thousands of others trying to buy the same shoes. After missing out on a chance to buy a pair of Yeezys for the hundreth time this Thanksgiving, I knew I needed to get a better grasp of the resale market. 

As of the summer of 2019, the sneaker resale market is supposed to be worth $2 billion dollars and growing, according to Cowen & Co. If you've been following the industry, you know what limited release sneakers are always the most sought after sneakers of the year. The only problem is sneaker companies such as Nike and Adidas will never release how many pairs of particular sneaker they are actually going to make and release. Therefore, I wondered what were some indicators of a sneaker that is going to be worth more money than they initially retail for.

For this analysis, I'm going to look at sneaker characteristics from the three most popular sneaker brands, Adidas, Air Jordan, and Nike, or the "Big 3", and understand what contributes to the resale value of a sneaker. The three characteristics this analysis will cover:

    1) Popularity
    2) Collaborations
    3) Colors
    
The data used for this analysis was scraped from three different websites. A sneaker database was created from the Kicks on Fire website, resale prices of sneakers was scraped from StockX and silhouette data were scraped from Sole Collector.

##### Key Terminology:
    - Silhouettes: A silhouette is the shape of a sneaker. For example, Air Jordan 1 is commonly known silhouette.
    - StockX: A "Stock market of things". A commonly used platform to buy and sell shoes.
    - Sole Collector: A sneaker website.
    - Kicks on Fire: A sneaker website.

## How many sneakers actually increase in resale value? 

Before trying to understand what contributes to sneaker resale value, I thought it would be important to know the distribution of gross profit for individial sneakers. The gross profit of each sneaker was calculated by taking the difference between resale value of a sneaker (at least 2 months after the sneaker is released -- that is around the time the resale price stabilizes) and it's retail value.

The violin and box plots below show the distribution of sneakers by their gross profit, split by their brands. Each individual point on the graph represents an individual sneaker. Most sneakers gross profit are centered around the break even line for each brand and very few sneakers stray away from their retail value.

## 1) Popularity

### There is no correlation between the gross profit and popularity of sneakers. Also, Jordan's maintain their value despite what people think of them.

Each sneaker on the Kicks on Fire website has a "Wants" score that is similar to a like on Facebook or Instagram. People press this button when they want a sneaker. For this analysis, I used this score as a proxy for sneaker popularity. Therefore, sneakers with a higher number of "Wants" would be considered more popular than those that have fewer.

The scatter plot below revealed that there was no real correlation between Kicks on Fire "Wants" and price difference. It also shows that Air Jordan sneakers retain value despite public opinion.

## 2) Collaborations

### The same silhouettes that were collaborated on usually sell for more than those that were not.

From Gucci, to Kanye to even Spongebob, sneaker companies have been collaborating with various people, companies and even TV shows in order to inspire sneakers design. To understand how collaborations effect sneaker resale prices, the difference between the average gross prices of silhouettes with collaborations versus those that didn't have ones were taken. The 9 silhouettes chosen in the bar chart are the three silhouettes that had the most number of appearances in the Kicks on Fire database from each of the "Big 3" brands.

As can be seen from the bar chart, collaborating on a sneaker can drastically increase the resale value of the sneaker. For example, Air Jordan 1's that have been collaborated on average are almost $550 more than regular Air Jordan 1's. Although the resale value of sneakers usually experience an increase in value after a collaboration, that is not always the case. For example, on average, the Air Jordan 12 actually lost value once collaborated on.

## 3) Colors

### If you want to resell your sneakers, don't buy the yellow ones.

Probably the thing that draws us the most to a sneaker at a store, is a sneaker's color. A sneaker can be all-white or all-black but it can also be multi-colored or a mix of yellow, green, and blue. When sneaker are released they are not described as "red" or "green", but usually as something like "Sail/Black-Habanero Red-Black" or "University Gold/Flash Crimson-Racer Blue". So what color is that? In order to define a "main color" for each sneaker, I used an NLP technique called "word2vec" to get the semantic similarity between these crazy color names to standard colors such as "black" or "blue". The color with the highest similarity score was then assigned to be the main color for that sneaker.

Although many of these distributions look the same, there is quite a bit of insight that can be drawn from them. On the y-axis is the percentage of sneakers that fall in each gross profit bucket (groups of 10 dollars). Most sneakers lie in the 0 - 10 dollar bucket. Some colors such as black, blue, red, and, white have a higher percentage of sneakers with higher gross profits. One distribution that is a significant outlier, is the yellow one. Most yellow sneakers are in the negative gross profit buckets and do not have a large spike in the 0 - 10 dollar buckets like the other distributions.

*Note: The grey distribution is actually a distribution for multi-color sneakers.*
