# Sneaker Resale Analysis: What causes some sneakers to increase in value after they're released?

As an avid "sneakerhead", I often find myself not being able to purchase the sneakers I truly want due to a limited supply of them and trying to compete with hundreds of thousands of people online trying to buy the same shoes. After missing out on a chance to buy a pair of Yeezys for the tenth time over the past three years, I wondered if it possible to know what sneakers are going to be resold for a premium. Although this is an obvious issue of supply and demand, sneaker companies such as Nike and Adidas never release how many pairs of particular sneaker they are going to drop. Therefore, I wondered if there are important characteristics of sneaker, outside of supply, that indicate a sneaker is going to resale for more money than they retail for, and thus be hard to get.

For this analysis, I'm going to look at three sneaker characteristics that I believe could help us understand what increases or decreases the value of a sneaker. They are:

    1) Public sentiment
    2) Collaborations with other companies, people, and movies/shows
    3) Colors
    
The data used for this analysis was scraped from three different websites. A sneaker database was created from the Kicks on Fire website, resale prices of sneakers was scraped from StockX and silhouette data were scraped from Sole Collector. 

## How many sneakers actually increase in resale value? 

Before trying to understand what controbutes to sneaker resale value, I thought it would be important to know the distribution of sneakers resale price in relation to their retale value. The price difference metric was created to represent the difference between the resale value of a sneaker and then retail value of a sneaker (resale price - retail price). Therefore, any data point below the breakeven point would be a sneaker that lost value after it was released and vice versa.

The violin and box plot below show that most sneakers price difference are centered around the break even point for each of the big 3 companies, and very few data points are in the extremes.

## 1) There is no correlation between sneaker sentiment and price difference.

Each sneaker on the Kicks on Fire website has a "Wants" score that is similar to a like on Facebook or Instagram. People press this button when they "want" a sneaker. This score was used as a proxy for consumer sentiment. Therefore, sneakers with a higher number of "Wants" would be considered more popular than those that have fewer.

The scatter plot below revealed that there was no real correlation between Kicks on Fire "Wants" and price difference. It did, however, visualize how Air Jordan sneakers retain value despite such a fluctuation of emotion.

## 2) The same silhouettes that were collaborated on usually sell for more than those that were not.

From Gucci, to Kanye to even Spongebob, sneaker companies have started collaborating with various people, companies and even TV shows in order to design sneakers. To understand how collaborations effect sneaker resale prices, the difference between the average price difference of silhouettes with collaborations versus those that didn't have ones were taken. The 9 silhouettes chosen in the bar chart are the 3 silhouettes that had the most number of appearances in the Kicks on Fire database from each of the big 3 companies.

As can be seen from the bar chart, collaborating on a sneaker can drastically increase the resale value of the sneaker. For example, on average, Air Jordan 1's that have been collaborated on are almost $550 more than regular Air Jordan 1's. Although the resale value of sneakers usually experience an increase in value after a collaboration, that is not always the case. For example, on average, the Air Jordan 12 actually lost value once collaborated on.

## 3) Certain colors have more potential for positive resale values than others.

Probably the one thing that draws us to a sneaker most when we go to a store, is a sneaker's color. A sneaker can be all-white or all-black but it can also be multi-colored or a mix of yellow, green, and blue. When sneaker colors are described on most websites, it usually will say something like "Sail/Black-Habanero Red-Black" or "University Gold/Flash Crimson-Racer Blue". So what color is that? In order to define a "main color" for each sneaker, I used an NLP technique called "word2vec" to get the semantic similarity between these crazy color names to standard colors such as black or blue. The color with the highest similarity score was then assigned to be the main color for that sneaker.

Although many of these distributions look the same, there is quite a bit of insight that can be drawn from them. On the y-axis is the percentage of sneakers that fall in each resale price premium bucket (groups of 10%). Most sneakers lie in the 0-10% bucket. Some colors such as black, blue, red, and, white have a higher percentage of sneakers with higher resale price premiums. One distribution that is a significant outlier are yellow sneakers. Most yellow sneakers have a negative resale price premium and do not have a large spike in the 0-10% price premium bucket like all of the other buckets.

*Note: The grey distribution is actually a distribution for multi-color sneakers.*

## Conclusion

In conclusion, all three characteristics of sneakers were insightful. Perhaps the most import takeaway was how much sneaker collaborations can increase the resale value of a sneaker. Although it was not always the case, there were often large increases to the resale value of collaborated silhouettes. Another critical insight was how poorly yellow sneakers did on the resale market. Around 70% of yellow sneakers experience a decrease in value once released.
